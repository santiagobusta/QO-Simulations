```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib.lines import Line2D
```

```python
N = 5000
M = 128

phi = np.linspace(0,2*np.pi,M)

C0 = np.load('coincidences_psi+_N{}_M{}.npy'.format(N,M))
C1 = np.load('coincidences_psi-_N{}_M{}.npy'.format(N,M))
C2 = np.load('coincidences_phi+_N{}_M{}.npy'.format(N,M))
C3 = np.load('coincidences_phi-_N{}_M{}.npy'.format(N,M))
```

```python
plt.figure()
plt.plot(phi, C0/N, lw = 2, label=r'$| \psi^+ \rangle$')
plt.plot(phi, C1/N, lw = 2, label=r'$| \psi^- \rangle$')
plt.plot(phi, C2/N, lw = 2, label=r'$| \phi^+ \rangle$')
plt.plot(phi, C3/N, lw = 2, label=r'$| \phi^- \rangle$')
plt.legend(loc='upper right', fontsize=9)
plt.grid()
plt.xticks([0,np.pi,2*np.pi], [r'$0$',r'$\pi$',r'$2\pi$'])
plt.ylabel('Normalized coincidences', fontsize=12)
plt.xlabel('Phase difference (rad)', fontsize=12)
plt.show()
```

```python
N = 500
M = 128
L = 30

phi = np.linspace(0,2*np.pi,M)

C0 = np.load('coincidences_psi+_N{}_M{}_L{}.npy'.format(N,M,L))
C1 = np.load('coincidences_psi-_N{}_M{}_L{}.npy'.format(N,M,L))
C2 = np.load('coincidences_phi+_N{}_M{}_L{}.npy'.format(N,M,L))
C3 = np.load('coincidences_phi-_N{}_M{}_L{}.npy'.format(N,M,L))
```

```python
plt.figure(figsize=(8,6))
for l in range(L):
    plt.scatter(phi, C0[:,l]/N, s=3, marker = '^', c = C0[:,l]/N, cmap= cm.autumn)
    plt.scatter(phi, C1[:,l]/N, s=3, marker = 'v', c = C1[:,l]/N, cmap= cm.winter)
    plt.scatter(phi, C2[:,l]/N, s=3, marker = 'd', c = C0[:,l]/N, cmap= cm.summer)
    plt.scatter(phi, C3[:,l]/N, s=3, marker = 's', c = C1[:,l]/N, cmap= cm.spring)
    
#plt.legend(loc='upper right', fontsize=9)
C0_label = Line2D([0], [0], marker='^', markersize=5, linestyle='', color = cm.autumn(0.5), label=r'$| \psi^+ \rangle$')
C1_label = Line2D([0], [0], marker='v', markersize=5, linestyle='', color = cm.winter(0.5), label=r'$| \psi^- \rangle$')
C2_label = Line2D([0], [0], marker='d', markersize=5, linestyle='', color = cm.summer(0.5), label=r'$| \phi^+ \rangle$')
C3_label = Line2D([0], [0], marker='s', markersize=5, linestyle='', color = cm.spring(0.5), label=r'$| \phi^- \rangle$')
plt.legend(handles = [C0_label,C1_label,C2_label,C3_label], loc='upper right')
plt.grid()
plt.xticks([0,np.pi/2.,np.pi,3*np.pi/2,2*np.pi], [r'$0$',r'$\pi/2$',r'$\pi$',r'$3\pi/2$',r'$2\pi$'])
plt.ylabel('Normalized coincidences', fontsize=12)
plt.xlabel('Phase difference (rad)', fontsize=12)
plt.savefig('NormalizedCoincidences_vs_PhaseDifference_N{}_M{}_L{}.jpeg'.format(N,M,L), dpi=800)
plt.show()
```
